```python
import src.cosmology as cos
import src.massfunc as mf
import numpy as np
import matplotlib.pyplot as plt

#Cosmo = cos.Cosmology()

zz = np.arange(0.0,2.01,0.1)

HMF = mf.HMF(version='col',zarr=zz)
HMF2 = mf.HMF(zarr=zz)

M = 10**np.arange( np.log10(5e13), 15.7, 0.01) #11.0, 15.5, 0.1)

mf_col = HMF.dn_dM(M*0.7,delta=500)
mf_int = HMF2.dn_dM(M,delta=500)
```

```python
plt.figure()
plt.xlabel('M500c')
plt.ylabel('dn/dln(M)')
plt.loglog()
plt.xlim(5E13, 7E15)
plt.ylim(1E-17, 1E-5)
plt.plot(M, mf_col[:,18] * M, label = 'Sculi')
plt.plot(M, mf_int[:,18] * M, '--',label = 'Sculi2')
plt.legend()

plt.figure()
plt.xlabel('M500c')
plt.ylabel('Ratio Colossus / SCuLi')
plt.xlim(3E13, 5E15)
for i in range(len(HMF.zarr)):
    plt.semilogx(M, mf_col[:,i]/mf_int[:,i])
#plt.legend()
```
